```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train = pd.read_table('input/train_20171215.txt', engine='python')
train= train.groupby(['date', 'day_of_week'], as_index=False)['cnt'].agg({'cnt': np.sum})
#train['cnt']=np.log1p(train['cnt'])
```

```python
#train1=train[train['date']<600]
#train1=train1[train1['date']<200]
print('全部数据')
train1=train
plt.plot(train1['date'],train1['cnt'])
plt.show()
```

```python
t0=150
train1=train[train['date']<=365+t0]
train1=train1[train1['date']>t0]
print('第一年')
plt.plot(train1['date'],train1['cnt'])
plt.show()
```

```python
train1=train[train['date']<=365*2+t0]
train1=train1[train1['date']>365+t0]
print('第2年')
plt.plot(train1['date'],train1['cnt'])
plt.show()
```

```python
train1=train[train['date']<=365*3+t0]
train1=train1[train1['date']>365*2+t0]
print('第3年')
plt.plot(train1['date'],train1['cnt'])
plt.show()
```
